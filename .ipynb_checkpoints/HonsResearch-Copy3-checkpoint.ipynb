### Forecasting the South African Unemployment Rate Using Deep Learning and Bayesian Techniques.

```python
import pandas as pd
import numpy as np
from numpy import sqrt
import matplotlib.pyplot as plt
import impyute as impy

from pylab import rcParams
import statsmodels.api as sm

import statsmodels
from statsmodels.tsa.seasonal import seasonal_decompose

import sklearn
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

### Problems with dataset

__Missing Data from mixing frequencies__: Due to varying frequencies being mixed there is some missing data. <br>
This is the result of merging monthly data and quarterly data. This is expected when working with different time series. <br>

Some data observations will be deleted accordingly to match the target variable to improve the percentage of missing data. <br>


```python
#Here we visualize the frequency of missingness of each feature
feature = features.isna().sum()/len(features)

feature = feature.sort_values(ascending=False)
feature_df = pd.DataFrame(feature.index, columns=['Feature Name'])
feature_df['Missing Frequency'] = np.array(feature.values)
feature_df['Rank'] = feature_df['Missing Frequency'].rank(ascending=False)
feature_df
```

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=90)
plt.bar(feature_df['Feature Name'], feature_df['Missing Frequency'])
```

### Upsampling Monthly Data to Quarterly Data

```python
df = features
df.reset_index(inplace=True)
df.Timestamp = pd.to_datetime(df.Date,format='%Y-%m-%d') 
df.index = df.Timestamp 
df = df.resample('Q').mean()
features = df
features.head()
```

```python
d = target
d.reset_index(inplace=True)
d.head()
d.Timestamp = pd.to_datetime(d.Date,format='%Y-%m-%d') 
d.index = d.Timestamp 
d = d.resample('Q').mean()
target = d
target.head()
```

__Last known value data imputation__ <br>


```python
# forward fill
x_values_ffill = features.fillna(method='ffill')
y_values_ffill = target.fillna(method='ffill')
```

```python
# remove all data points before unemployment rate data is available 
# unemployment rate is the target variable

valid_start = y_values_ffill.first_valid_index()
y_values_ffill = y_values_ffill[valid_start :]
x_values_ffill = x_values_ffill[valid_start :]
```

```python
# fill with NA here to avoid any features that might be NA i.e. insurance
x_values_ffill = x_values_ffill.fillna(0)
x_values_ffill.isna().sum()
```

```python
rcParams['figure.figsize'] = 20,10
decompose_result = seasonal_decompose(y_values_ffill['unemployment rate'], model='add', period=12)
fig = decompose_result.plot()
plt.show()
```

__Expectation Maximization__ <br>

It is an iterative procedure in which it uses other variables to impute a value (Expectation), 
then checks whether that is the value most likely (Maximization).

```python
# remove all data points before unemployment rate data is available 
# unemployment rate is the target variable

valid_start = target.first_valid_index()
features2 = features[valid_start :]
target2 = target[valid_start :]
```

### Long Short-Term Memory

```python
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf

import keras
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.layers import LSTM, GRU, RNN, Embedding
from keras.callbacks import EarlyStopping
from keras.wrappers.scikit_learn import KerasRegressor

import scikeras
from scikeras.wrappers import KerasRegressor

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# normalize the dataset - FF
scaler = MinMaxScaler(feature_range=(0, 1))

LSTMX_train = ffillX_train.values
LSTMX_train = scaler.fit_transform(LSTMX_train)

LSTMY_train = ffillY_train.values.reshape(-1, 1)
LSTMY_train = scaler.fit_transform(LSTMY_train)

LSTMX_test = ffillX_test.values
LSTMX_test = scaler.fit_transform(LSTMX_test)

LSTMY_test = ffillY_test.values.reshape(-1, 1)
LSTMY_test = scaler.fit_transform(LSTMY_test)

# reshape input to be [samples, time steps, features]
LSTMX_train = np.reshape(LSTMX_train, (LSTMX_train.shape[0], 1, LSTMX_train.shape[1]))
LSTMX_test = np.reshape(LSTMX_test, (LSTMX_test.shape[0], 1, LSTMX_test.shape[1]))
```

```python
# normalize the dataset - EM
emLSTMX_train = EMX_train
emLSTMX_train = scaler.fit_transform(emLSTMX_train)

emLSTMY_train = EMY_train.values.reshape(-1, 1)
emLSTMY_train = scaler.fit_transform(emLSTMY_train)

emLSTMX_test = EMX_test.values
emLSTMX_test = scaler.fit_transform(emLSTMX_test)

emLSTMY_test = EMY_test.values.reshape(-1, 1)
emLSTMY_test = scaler.fit_transform(emLSTMY_test)

# reshape input to be [samples, time steps, features]
emLSTMX_train = np.reshape(emLSTMX_train, (emLSTMX_train.shape[0], 1, emLSTMX_train.shape[1]))
emLSTMX_test = np.reshape(emLSTMX_test, (emLSTMX_test.shape[0], 1, emLSTMX_test.shape[1]))
```

```python
def create_model(learning_rate):
    
    model = Sequential()
    model.add(LSTM(555, activation='relu', dropout = 0.2, input_shape=(1, 147)))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam')
    
    return model
```

```python
model = create_model(learning_rate = 0.1)
model.compile()
model = KerasRegressor(model = create_model, 
                     verbose = 0
                     )
```

__Training and Testing__ - forward fill data set

```python
lstm = create_model(learning_rate = 0.1)
lstm.fit(LSTMX_train, LSTMY_train, epochs = 100, batch_size = 10)
```

```python
expected_lstmy = LSTMY_test
predicted_lstmy = lstm.predict(LSTMX_test)
print("mean_absolute_error : ", metrics.mean_absolute_error(expected_lstmy, predicted_lstmy))
MSE = mean_squared_error(expected_lstmy, predicted_lstmy)
print("rmse : ", sqrt(MSE))
```

__Training and Testing__ - Expectation Maximization data set

```python
lstmEM = create_model(learning_rate = 0.1)
lstmEM.fit(LSTMX_train, LSTMY_train, epochs = 100, batch_size = 10)
```

```python
expected_lstmy = emLSTMY_test
predicted_lstmy = lstmEM.predict(emLSTMX_test)
print("mean_absolute_error : ", metrics.mean_absolute_error(expected_lstmy, predicted_lstmy))
MSE = mean_squared_error(expected_lstmy, predicted_lstmy)
print("rmse : ", sqrt(MSE))
```

### Gated Recurrant Networks

__Training and Testing__ - forward fill data set

```python
def create_gru(learning_rate):
    
    model = Sequential()
    model.add(GRU(555, activation='relu', dropout = 0.2, input_shape=(1, 147)))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(10))
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam')
    
    return model

model = create_gru(learning_rate = 0.1)
model.compile()
model = KerasRegressor(model = create_gru, 
                     verbose = 0
                     )

gru = create_gru(learning_rate = 0.1)
gru.fit(LSTMX_train, LSTMY_train, epochs = 100, batch_size = 10)
```

```python
expected_gru = LSTMY_test
predicted_gru = gru.predict(LSTMX_test)
print("mean_absolute_error : ", metrics.mean_absolute_error(expected_gru, predicted_gru))
MSE = mean_squared_error(expected_gru, predicted_gru)
print("rmse : ", sqrt(MSE))
```

__Training and Testing__ - Expectation Maximization data set

```python
expected_gruEM = emLSTMY_test
predicted_gruEM = gruEM.predict(emLSTMX_test)
print("mean_absolute_error : ", metrics.mean_absolute_error(expected_gruEM, predicted_gruEM))
MSE = mean_squared_error(expected_gruEM, predicted_gruEM)
print("rmse : ", sqrt(MSE))
```

### Holt-Winters

### AutoRegressive Integrated Moving Average
