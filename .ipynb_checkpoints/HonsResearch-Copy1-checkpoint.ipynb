### Problems with dataset

__Missing Data from mixing frequencies__: Due to varying frequencies being mixed there is some missing data. <br>
This is the result of merging monthly data and quarterly data. This is expected when working with different time series. <br>

Some data observations will be deleted accordingly to match the target variable to improve the percentage of missing data. <br>


```python
#Here we visualize the frequency of missingness of each feature
feature = features.isna().sum()/len(features)

feature = feature.sort_values(ascending=False)
feature_df = pd.DataFrame(feature.index, columns=['Feature Name'])
feature_df['Missing Frequency'] = np.array(feature.values)
feature_df['Rank'] = feature_df['Missing Frequency'].rank(ascending=False)
feature_df
```

It is an iterative procedure in which it uses other variables to impute a value (Expectation), 
then checks whether that is the value most likely (Maximization).

__Bayesian Neural Networks__ <br>
Building a probabilistic Bayesian neural network models with TensorFlow Probability.

```python
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
import tensorflow_probability as tfp
import tensorflow_datasets as tfds
```

```python
def prior(kernel_size, bias_size, dtype=None):
    n = kernel_size + bias_size
    prior_model = keras.Sequential(
        [
            tfp.layers.DistributionLambda(
                lambda t: tfp.distributions.MultivariateNormalDiag(
                    loc=tf.zeros(n), scale_diag=tf.ones(n)
                )
            )
        ]
    )
    return prior_model
```

```python
def posterior(kernel_size, bias_size, dtype=None):
    n = kernel_size + bias_size
    posterior_model = keras.Sequential(
        [
            tfp.layers.VariableLayer(
                tfp.layers.MultivariateNormalTriL.params_size(n), dtype=dtype
            ),
            tfp.layers.MultivariateNormalTriL(n),
        ]
    )
    return posterior_model
```
